# Scraping from CarTrade Website

```python
# activating the chrome browser
driver=webdriver.Chrome("C:/Users/jprat/Downloads/chromedriver_win32 (1)/chromedriver.exe")
```

```python
# opening used cars page on website
driver.get('https://www.cartrade.com/buy-used-cars/#sc=-1&so=-1&pn=1')
time.sleep(8)
# closing pop-up
close=driver.find_element_by_xpath("//a[@id='citybandclose']")
close.click()
```

```python
# extracting used cars url
cars_url = []
for page in range(0,100): # extracting used cars url from page1 to page 100
    for i in driver.find_elements_by_xpath("//h2[@class='h2heading truncate']/a"):
        cars_url.append(i.get_attribute('href'))
    try: # to enter next page   
        nxt_page=driver.find_element_by_xpath("//li[@class='next']/a").get_attribute('href')
        driver.get(nxt_page)
        delay=8
        WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH,"//a[@id='citybandclose']")))
        close=driver.find_element_by_xpath("//a[@id='citybandclose']") # closing pop-up
        close.click()
        delay=8
        WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH,"//h2[@class='h2heading truncate']/a")))
    except NoSuchElementException:
        break
```

```python
len(cars_url)
```

```python
%%time

# Let's create empty lists to store necessary information
brand_model1 = []
fuel1 = []
kms_driven1 = []
no_of_owners1 = []
make_year1 = []
price1 = []

for url in cars_url[:600]: # extracting cars details in chunks
    driver.get(url) # opening the car page
    try:
        delay=10
        WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH, "//h1[@class='h_2']")))
    except:
        continue
    
#     extracting brand and model
    try:
        bm=driver.find_element_by_xpath("//h1[@class='h_2']")
        brand_model1.append(bm.text.replace('Used ',''))
    except NoSuchElementException:
        brand_model1.append(np.NaN)
    
#     extracting FUEL TYPE
    try:
        for i in driver.find_elements_by_xpath("//table[@class='v_table']/tbody/tr[2]/td[2]"):
            fuel1.append(i.text)
    except NoSuchElementException:
        fuel1.append(np.NaN)
    
#     extracting KMS Driven
    try:
        for i in driver.find_elements_by_xpath("//table[@class='v_table']/tbody/tr[3]/td[2]"):
            kms_driven1.append(i.text.replace(' Kms',''))
    except NoSuchElementException:
        kms_driven1.append(np.NaN)
    
#     extracting NUMBER OF OWNERS
    try:
        for i in driver.find_elements_by_xpath("//table[@class='v_table']/tbody/tr[5]/td[2]"):
            no_of_owners1.append(i.text)
    except NoSuchElementException:
        no_of_owners1.append(np.NaN)
    
#     extracting MANUFACTURING YEAR
    try:
        for i in driver.find_elements_by_xpath("//table[@class='v_table']/tbody/tr[8]/td[2]"):
            make_year1.append(i.text)
    except NoSuchElementException:
        make_year1.append(np.NaN)
    
#     extracting price
    try:
        for i in driver.find_elements_by_xpath("//div[@class='pull-left']/span[2]"):
            price1.append(i.get_attribute('content'))
    except NoSuchElementException:
        price1.append(np.NaN)
```

```python
%%time

# Let's create empty lists to store necessary information
brand_model1 = []
fuel1 = []
kms_driven1 = []
no_of_owners1 = []
make_year1 = []
price1 = []

for url in cars_url[:500]: # extracting cars details in chunks
    driver.get(url) # opening the car page
    try:
        delay=10
        WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH, "//h1[@class='d-font-weight-500 d-font-size-22 d-text-dark d-margin-top-0']")))
    except:
        continue
#     extracting brand and model
    try:
        bm=driver.find_element_by_xpath("//h1[@class='d-font-weight-500 d-font-size-22 d-text-dark d-margin-top-0']")
        brand_model1.append(bm.text)
    except NoSuchElementException:
        brand_model1.append(np.NaN)
    
#     extracting FUEL TYPE, KMS Driven, NUMBER OF OWNERS    
    info = []
    try:
        for i in driver.find_elements_by_xpath("//div[@class='d-margin-top-2 d-text-gray d-font-size-12']"):
            info.append(i.text)
        fuel1.append(info[3])            # extracting FUEL TYPE
        kms_driven1.append(info[1])      # extracting KMS Driven
        no_of_owners1.append(info[0])    # extracting NUMBER OF OWNERS
    except NoSuchElementException:
        fuel1.append(np.NaN)            
        kms_driven1.append(np.NaN)      
        no_of_owners1.append(np.NaN)
    
#     extracting MANUFACTURING YEAR    
    try:
        my=driver.find_element_by_xpath("//h1[@class='d-font-weight-500 d-font-size-22 d-text-dark d-margin-top-0']")
        make_year1.append(my.text.split()[-1])
    except NoSuchElementException:
        make_year1.append(np.NaN)
    
#     extracting price    
    try:
        p=driver.find_element_by_xpath("//div[@class=' d-font-size-22']")
        price1.append(p.text.replace('₹ ','').replace(',',''))    
    except NoSuchElementException:
        p=driver.find_element_by_xpath("//div[@class=' d-text-decoration-line-through  d-font-size-22']")
        price1.append(p.text.replace('₹ ','').replace(',',''))
    except:
        price1.append(np.NaN)
```
